```python
df=pd.read_csv("census.csv")
df
```

```python
df=df[df['SUMLEV']==50]
```

```python
df.head()
```

```python
%%timeit -n 3
for state in df['STNAME'].unique():
    avg=np.average(df.where(df['STNAME']==state).dropna()['POPESTIMATE2010'])
    print("Countires in sate "+state+"have a average population"+str(avg))
```

```python
%%timeit -n 3
# now by using groupby function
for group,frame in df.groupby('STNAME'):
#     groupy returns a tuple of two values 
#value of the key we are trying to group by inthis a specific state name and the second one is the projected dataframe that is found for this group 
    avg=np.average(frame['POPESTIMATE2010'])
    print("avd =",avg)
        

   
```
