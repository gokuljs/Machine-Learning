```python
import pandas as pd 
```

```python
df=pd.read_csv("presidents.csv")
df.head()
```

```python
df=df.rename(mapper=str.lower,axis="columns")
df=df.rename(mapper=str.strip,axis="columns")
df
```

```python
def splitname(row):
#     the row is a single series object whihc is single row indexed by column name
    row['first']=row['president'].split(" ")[0]
    row['last']=row['president'].split(" ")[-1]
    return row

df=df.apply(splitname,axis='columns')
df.head() 
```
